```python
import numpy as np
import pandas as pd


train = pd.read_csv('./data/train.csv', encoding='big5')
train.head()
```

```python
train = train.sort_values(['日期', '測項'], ascending = [True, True])
train.head()
```

```python
train_data = train.copy()

PM25_only = True
if PM25_only: 
    train_data = train_data[train_data['測項'] == 'PM2.5']
train_data = train_data.iloc[:, 3:]
train_data[train_data == 'NR'] = 0
train_data.head()
```

```python
train_numpy = train_data.to_numpy()
train_numpy.shape
```

```python
month_data = {}

n_month, n_day, n_hour = 12, 20, 24
for month in range(n_month):
    sample = np.empty([1, n_day * n_hour], dtype = float)
    for day in range(n_day):
        sample[:, day * 24 : (day + 1) * 24] = train_numpy[month * 20 + day: month * 20 + day + 1, :]
    month_data[month] = sample
    
month_data.keys()
```

```python
month_data[0]
```

```python
time_range = 9

x = np.zeros([(n_day * 24 - time_range) * n_month, time_range], dtype = float)
y = np.zeros([(n_day * 24 - time_range) * n_month, 1], dtype = float)

index = 0
for month in range(n_month):
    for i in range(month_data[month].shape[1] - 9):
        x[index: index + 1, :] = month_data[month][:, i: i + 9]
        y[index: index + 1, :] = month_data[month][:, i + 9: i + 10]
        index += 1
    

x.shape, y.shape
```

## Normalization

```python
mean_x = np.mean(x)
std_x = np.std(x)

norm_x = (x - mean_x) / std_x
```

# Train

### split into train / validation

```python
n, percentage = len(x), 0.8
train_x, val_x = norm_x[: int(n * 0.8)], norm_x[int(n * 0.8):]
train_y, val_y = y[: int(n * 0.8)], y[int(n * 0.8):]

print(train_x.shape, train_y.shape)
print(val_x.shape, val_y.shape)
```

```python
# add bias
train_X = np.concatenate((train_x, np.ones((train_x.shape[0], 1))), axis = 1)
# init weight
w = np.random.random((train_X.shape[1], 1))

learning_rate = 100
iter_time = 10000
adagrad = np.zeros((train_X.shape[1], 1))
eps = 10 ** -10
for iter_ in range(iter_time):
    loss = np.sqrt(np.sum(np.power(np.dot(train_X, w) - train_y, 2)) / train_X.shape[0]) # Root Mean Square Error
    
    if(iter_ % 1000 == 0):
        print('{iter_} : {loss}'.format(iter_ = iter_, loss = loss))
        
    gradient = 2 * np.dot(train_X.transpose(), np.dot(train_X, w) - train_y) # dim * 1
    adagrad += gradient ** 2
    w = w - learning_rate * gradient / np.sqrt(adagrad + eps)
np.save('weight.npy', w)
```
